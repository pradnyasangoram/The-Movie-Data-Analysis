```python
#droping unnessary columns
df= df.drop(['id','imdb_id', 'homepage', 'tagline', 'keywords', 'overview', 'budget_adj', 'revenue_adj'], axis=1)
```

```python
df['budget'].fillna(df['budget'].mean(), inplace=True)
df['revenue'].fillna(df['revenue'].mean(), inplace=True)
# Dropping null/NaN values in rows using df.dropna() function
df.dropna(axis=0, inplace=True)
```

```python
df
```

```python
# converting non-date data-type to date-type 
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = pd.DatetimeIndex(df['release_year'])
df
```

Expolatory Data Analysis:

##### Q1. What are the highest grossing films of all the time?

```python
fig, ax = plt.subplots(figsize= (12, 6))
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 12
high_gross = df.sort_values(by='revenue', ascending= False)[:10]
plt.bar(high_gross['original_title'], high_gross['revenue'])
plt.xlabel('Film')
plt.ylabel('Revenue')
plt.title('heighest Grossing Films of all time')
plt.xticks(rotation=90)
plt.show()
```

Avatar(2.78B) was highest grossing film of all time, followed by Star Wars(2.06B) and Titanic(1.84B).



##### Q2. Which day of the week has most number of releases?

```python
fig, ax = plt.subplots(figsize = (12,6))
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size']=12
day_week = df['release_date'].dt.day_name()
day_week.value_counts().plot.bar()
plt.xticks(rotation=45)
plt.xlabel('Weekday')
plt.ylabel('Films')
plt.title('Most Films by Day of Week')
plt.show()
```

Friday seems to have most of the releases followed by thursday and wednesday.

##### Q3. How does popularity influence revenue of a film?

```python
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size']=12
df.plot(x='popularity', y='revenue', kind='scatter', figsize=(12,6))
plt.title('Popularity vs Revenue')
plt.show()
```

There is General positive correlation among popularity and revenue with few outliers.
